# MagCAMB test

This notebook provides a few example of how to use MagCAMB.

To compile the code run

- make clean
- make camb

Throughout this notebook we call "standard" the magnetic parameters with the following values

$$
B_{1 {\rm Mpc}} = 4.5 \, {\rm nG}, \quad n_{B} = -2.9, \quad \log_{10}\left( \frac{\tau_{\nu}}{\tau_{B}}\right) = 17.
$$



```python
# importing the libraries to make some plots
import numpy as np
import matplotlib.pyplot as plt
import pylab as pil
```

## Scalar Compensated Modes

Here we show the scalar compensated modes contribution to the CMB anisotropies from the "standard" PMF parameters defined above.

To calculate the scalar compensated modes set 

- get_scalar_cls = T
- get_vector_cls = F
- get_tensor_cls = F
- get_transfer   = F
- magnetic_mode = 1

in params_mag.ini.

Then run the code with 

- ./camb params_mag.ini

The output will be 

some output_root_scalCls.dat according to CAMB standard.

In the following we show the outputs for the scalar compensated modes. 


```python
# importing the data
cls_sc_comp = np.loadtxt('./benchmark/test_SC_comp_scalCls.dat')
cls_primary = np.loadtxt('./benchmark/testGR_scalCls.dat')
```

```python
cls_sc_comp.shape
```

```python
# plotting the CMB power spectra
plt.loglog(cls_sc_comp[:,0], cls_sc_comp[:,1], label='Sc.Comp.')
plt.loglog(cls_primary[:,0], cls_primary[:,1],label='primary')
plt.xlim(2,2500)
plt.legend(loc='lower right')
plt.xlabel(r'$\ell$')
plt.ylabel(r'$\ell (\ell+1) C_{\ell}^{\rm TT} / 2 \pi \quad [\mu {\rm K}^2]$')
plt.show()
```

Fig.1: Scalar Compensated modes (blue line) and primary (orange line) contributions to the CMB temperature self-correlation anisotropies.

## Scalar and Tensor Passive Modes

The scalar and tensor passive modes can be calculated in one go (because they are the same transfer functions as in the primary anisotropies).

Set 

- get_scalar_cls = T
- get_vector_cls = F
- get_tensor_cls = T
- get_transfer   = F
- magnetic_mode = 2

in params_mag.ini.

and then run the code with 

- ./camb params_mag.ini

```python
# loading the data
cls_sc_pass = np.loadtxt('./benchmark/test_pass_scalCls.dat')
cls_tens_pass = np.loadtxt('./benchmark/test_pass_tensCls.dat')
```

```python
# plotting the CMB power spectra
plt.loglog(cls_sc_pass[:,0], cls_sc_pass[:,1], label='Sc.Pass.')
plt.loglog(cls_tens_pass[:,0], cls_tens_pass[:,1], label='Tens. Pass.')
plt.loglog(cls_primary[:,0], cls_primary[:,1],label='primary')
plt.xlim(2,2500)
plt.legend(loc='lower left')
plt.xlabel(r'$\ell$')
plt.ylabel(r'$\ell (\ell+1) C_{\ell}^{\rm TT} / 2 \pi \quad [\mu {\rm K}^2]$')
plt.show()
```

Fig.2: CMB anisotropies contribution from the scalar passive (blue), tensor passive (orange) and primary (green) modes. Note how the passive modes have exactly the same shape as the primary anisotropies.

## Vector Compensated Modes

The last contribution comes from the vector compensated modes. Here we show the contributions to the CMB TT anisotropies.

To calculate the vector compensated modes set

- get_scalar_cls = F
- get_vector_cls = T
- get_tensor_cls = F
- get_transfer   = F
- magnetic_mode = 1

in params_mag.ini.

and then run the code with 

- ./camb params_mag.ini

```python
cls_vec_comp = np.loadtxt('./benchmark/test_VecComp_vecCls.dat')
```

```python
plt.loglog(cls_vec_comp[:,0], cls_vec_comp[:,1], label='Vec.Comp.')
plt.loglog(cls_primary[:,0], cls_primary[:,1],label='primary')
plt.xlim(2,2500)
plt.legend(loc='lower right')
plt.xlabel(r'$\ell$')
plt.ylabel(r'$\ell (\ell+1) C_{\ell}^{\rm TT} / 2 \pi \quad [\mu {\rm K}^2]$')
plt.show()
```

Fig.3: contribution to the CMB TT anisotropies from the vector compensated (blue) and primary (orange) modes. Note how the vector modes contribute more at larger $\ell$s (smaller scales).
